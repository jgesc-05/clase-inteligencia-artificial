#Árboles de decisión

* Basado en condiciones, a diferencia de los otros modelos
* Es "caja blanca", a diferencia de la red neuronal, que es "caja negra" (no se puede saber cómo toma la decisión).

* El nodo raíz es una característica, que se tiene que binarizar (ejemplo, menores de 20 y mayores de 20)

* Se tiene que decidir el factor que queda en el nodo

* Es más puro mientras haya más de una clase en un nodo (se vuelve hoja)

* No se tiene que escalar y puede trabajar con variables categóricas sin problema.

* No son lineales, porque no usan ninguna fórmula matemática

* Se usan las mínimas variables requeridas para la decisión; los árboles de decisión tiene selección automática de features, por lo que no se debe usar manualmente.

* Muy propenso al overfitting

* Gradient boosting: los mismos árboles que aprenden de los árboles anteriores

* Gradientboosting classifier: mejores modelos

* El índice de 0 en un gini es que un nodo es puro, y un 1 es perfectamente impuro.

* El gini mas alto se coloca en la raiz, mientras que el gini mas bajo se usa en las últimas decisiones (mientras se desciende en el árbol, más bajo es el ginni)

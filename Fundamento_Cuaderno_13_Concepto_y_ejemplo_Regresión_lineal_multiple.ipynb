#<font color="redred">Cuaderno 13. Regresión lineal multiple

La regresión lineal es una técnica estadística y de machine learning utilizada para modelar la relación entre una variable dependiente o respuesta (Y) y una o más variables independientes o predictores (X). Este módulo explica la teoría, las métricas de evaluación, las condiciones necesarias, y un ejemplo completo implementado en Python.

En la imagen se muestra la "línea recta" que mejor se ajusta a los datos. Esta línea se describe con la ecuación:

$y=mx+b$

y es la variable dependiente (la que queremos predecir).

x es la variable independiente (la que usamos para predecir y).

m es la pendiente de la línea (indica cómo cambia y con respecto a 𝑥).

b es la intersección con el eje y (es el valor de y cuando 𝑥=0 ).

## <font color="rede">13.1 Definiciones

¿Qué es la Regresión Lineal múltiple?
La Regresión Lineal múltiple es un modelo que busca encontrar una relación lineal entre variables independientes (X) y la variable dependiente (Y).
Vamos a modificar las variables para reprensentar la Regresión Lineal Simple y adoptarla al modelo general múltiple, la siguiente escuación es igual que la anterior solo que usamos $ \beta$

#### Fórmula matemática:
$Y = \beta_0 + \beta_1 X + \epsilon$

### Descripción de los términos:
* **Y**: Variable dependiente.
* **X**: Variable independiente.
* **$beta_0$**: Intersección (valor de \( Y \) cuando \( X = 0 \)).
* **$beta_1$**: Pendiente (grado de cambio en \( Y \) por unidad de cambio en \( X \)).
* **$epsilon$**: Término de error.

### <font color="red">13.1.1 Ecuación con la Regresión Lineal Múltiple
Es una extensión de la regresión lineal simple, donde la variable dependiente está relacionada con dos o más variables independientes $(X_1, X_2, ..., X_p)$

Fórmula matemática:

  $Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \beta_p X_p + \epsilon$

Donde:  
* $Y$: Variable dependiente.  
* $X_1, X_2, \dots, X_p$: Variables independientes.  
* $\beta_0, \beta_1, \dots, \beta_p$: Coeficientes del modelo.

## <font color="rede">13.2 Bondad de Ajuste del Modelo
Error Estándar Residual (RSE)
El Error Estándar Residual mide la desviación promedio entre los valores observados y los valores predichos por el modelo.

$RSE = \sqrt{\frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{n - p - 1}}$

Donde:  
* $y_i$: Valor observado.  
* $\hat{y}_i$: Valor predicho.  
* $n$: Número de observaciones.  
* $p$: Número de predictores.  


### <font color="red">13.2.1 Coeficiente de Determinación ($R^2$)
El $R^2$ indica qué proporción de la variabilidad de la variable dependiente es explicada por el modelo.  

$$R^2 = 1 - \frac{\text{Suma de Cuadrados Residuales (RSS)}}{\text{Suma Total de Cuadrados (TSS)}}$$  

También puede calcularse como:  

$$R^2 = \frac{\sum (\hat{y}_i - \bar{y})^2}{\sum (y_i - \bar{y})^2}$$  

* Valor de $R^2$ cercano a 1 indica un buen ajuste.  

#### <font color="red">Significancia de los Predictores  

Se evalúa la significancia de los predictores usando el valor $p$ de las pruebas de hipótesis. Si $p < \alpha$ (nivel de significancia, comúnmente 0.05), el predictor es significativo.

## <font color="rede">13.4 Validación del Modelo
Para garantizar la validez de una regresión lineal, se deben verificar las siguientes condiciones:

### <font color="red">13.4.1 No Colinealidad o Multicolinealidad  

La multicolinealidad ocurre cuando los predictores en un modelo de regresión están altamente correlacionados entre sí, lo que dificulta la estimación precisa de los coeficientes. En presencia de multicolinealidad, los resultados del modelo pueden ser inestables, y las interpretaciones de los coeficientes pueden volverse poco confiables. Es importante evaluar y detectar la multicolinealidad en modelos de regresión.  

Para evaluar la multicolinealidad, se utilizan dos métodos principales:

***Índice de Tolerancia:*** Este índice mide cuánto de la varianza de un predictor puede explicarse por los demás predictores. Se calcula como:

$\text{Índice de Tolerancia} = 1 - R_j^2$

donde $(R_j^2)$ es el coeficiente de determinación de la regresión del predictor $j$ sobre los otros predictores. Un valor de tolerancia cercano a 0 indica alta multicolinealidad.

**Factor de Inflación de la Varianza (VIF, por sus siglas en inglés)**: El VIF mide cuánta varianza en el coeficiente estimado de un predictor se "infla" debido a la correlación con otros predictores. Se calcula como:
   
$VIF = \frac{1}{1 - R_j^2}$

donde $(R_j^2)$ es el coeficiente de determinación de la regresión del predictor $(j)$ sobre los demás predictores. Un VIF alto (típicamente mayor a 10) indica una alta multicolinealidad.

En general:
- Si la tolerancia es pequeña o el VIF es grande, significa que hay multicolinealidad.
- Se recomienda tratar de eliminar o combinar variables altamente correlacionadas para reducir la multicolinealidad y mejorar la estabilidad del modelo.

### <font color="red">13.4.2 Relación Lineal
La relación entre los predictores y la variable dependiente debe ser lineal. Puede verificarse mediante:
* Diagramas de dispersión
* Transformaciones logarítmicas si es necesario.

### <font color="red">13.4.3 Normalidad de los Residuos
Los residuos deben seguir una distribución normal. Puede verificarse con:
* Prueba de Shapiro-Wilk
* Gráficos Q-Q

### <font color="red">13.4.4 Homocedasticidad
La **homocedasticidad** se refiere a que la varianza de los residuos debe ser constante a lo largo de los valores ajustados (predicciones). Si la varianza de los residuos cambia con los valores ajustados, se dice que hay **heterocedasticidad**. Para verificar la homocedasticidad, se puede graficar los **residuos versus los valores ajustados**. En una distribución homogénea, los residuos deben estar dispersos aleatoriamente alrededor de cero, sin formar patrones sistemáticos. En los ejercicios prácticos vamos a ver la gráfica qqplot para verificar la varianza de los residuos.

### <font color="red">13.4.5 No Autocorrelación
La **autocorrelación** ocurre cuando los residuos no son independientes entre sí, es decir, hay correlación temporal o secuencial entre ellos. Esto es un problema porque violaría uno de los supuestos fundamentales de la regresión.

La **prueba de Durbin-Watson** se utiliza para evaluar la autocorrelación de los residuos. El estadístico de Durbin-Watson (\(d\)) se calcula con la siguiente fórmula:

$d = \frac{\sum_{t=2}^n (e_t - e_{t-1})^2}{\sum_{t=1}^n e_t^2}$

- $( e_t )$ es el residuo en el tiempo \(t\).
- $( d )$ varía entre 0 y 4:
  - $( d \approx 2 )$: No hay autocorrelación.
  - $( d < 2 )$: Autocorrelación positiva.
  - $( d > 2 )$: Autocorrelación negativa.

### <font color="red"><font color="rede">13.4.6 Ausencia de Valores Atípicos o Influyentes
Los valores atípicos pueden identificarse usando:
* Distancia de Cook
* Leverage

#### Paso 1: Importar librerías y cargar datos




Vamos a crear un DataFrame de pandas con datos de ejemplo. Cada columna representa una variable en el modelo:

* ***Edad:*** Edad de una persona.
* ***Ingresos:*** Ingresos anuales de una persona.
* ***Experiencia:*** Años de experiencia laboral de una persona.
* ***Satisfacción:*** Nivel de satisfacción de una persona (probablemente en una escala de 1 a 10).
* ***Compra:*** El valor de la compra realizada por la persona, que es la variable dependiente (es decir, la que queremos predecir o label) en este caso.

````python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Cargar datos de ejemplo
data = pd.DataFrame({
    'Edad': [25, 30, 35, 40, 45, 50, 55],
    'Ingresos': [40000, 50000, 60000, 80000, 100000, 110000, 120000],
    'Experiencia': [1, 3, 5, 7, 9, 11, 13],
    'Satisfacción': [3, 4, 5, 6, 8, 9, 10]
})
data['Compra'] = [10000, 12000, 15000, 18000, 20000, 22000, 25000]
````

````python
data
````

````python
data.describe().T
````

#### Paso 2: Dividir datos en conjuntos de entrenamiento y prueba

X: Contiene las variables independientes (Edad, Ingresos, Experiencia, Satisfacción).

y: Contiene la variable dependiente, en este caso, Compra.

Dado que son muy pocos datos no vamos vamos a dividir en entrenamento y prueba. Usaremos los mismos datos de entrenamiento para calculos de los residuos.

````python
# Características y variable respuesta
X = data[['Edad', 'Ingresos', 'Experiencia', 'Satisfacción']]
y = data['Compra']
````

````python
len(X), len(y)
````

#### Paso 3: Normalizar los datos


StandardScaler: Esta herramienta estandariza los datos, lo que significa que los valores de cada variable se ajustan para tener media cero y desviación estándar igual a uno. Esto es importante para la regresión lineal, ya que las variables en diferentes escalas (por ejemplo, ingresos en miles y edad en años) pueden afectar el rendimiento del modelo.

fit_transform: Ajusta el scaler a los datos de entrenamiento y los transforma para que tengan media cero y desviación estándar uno.

transform: Se utiliza para transformar los datos de prueba con los mismos parámetros que se ajustaron a los datos de entrenamiento.

````python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X)
````

````python
X_train_scaled=pd.DataFrame(X_train_scaled,columns=X.columns)
````

````python
X
````

````python
X_train_scaled
````

Dado que son muy pocos datos no vamos a hacer análisis de las variables o features desde los métodos. Vamos a validar solo validar las condiciones
* Multicolinealidad: Calcular VIF.
* Normalidad: Usar Q-Q plot o prueba de Shapiro-Wilk.
* Homocedasticidad: Gráfico de residuos.
* Independencia: Prueba de Durbin-Watson.
* Valores atípicos: Distancia de Cook.


#### Paso 4: Entrenar el modelo
LinearRegression: Aquí estamos creando un modelo de regresión lineal que intenta encontrar la relación entre las variables independientes y la variable dependiente.

fit: Este método entrena el modelo con los datos de entrenamiento (X_train_scaled y y_train).

````python
# Crear y entrenar modelo
model = LinearRegression()  #Declarar el modelo
model.fit(X_train_scaled, y) #Entrenar

# Coeficientes y ordenada al origen
print("Coeficientes:", model.coef_)
print("Intercepto:", model.intercept_)
````

#### Paso 5: Evaluar el modelo
predict: Una vez que el modelo ha sido entrenado, realizamos predicciones utilizando los datos de tentrenamento dado que son muy datos, lo normal es hacerlo con los datos de prueba (X_train_scaled).

mean_squared_error: Esta función calcula el error cuadrático medio (MSE), que mide la diferencia entre los valores predichos por el modelo y los valores reales. Un MSE más bajo indica que el modelo está realizando mejores predicciones.

r2_score: Este valor es el coeficiente de determinación (R²), que mide qué tan bien se ajusta el modelo a los datos. Un valor de R² cercano a 1 indica que el modelo explica bien la variabilidad de los dato

````python
# Predicciones
y_pred = model.predict(X_train_scaled)

y_pred
````

````python
np.array(y)
````

````python
# Métricas de evaluación
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("MSE:", mse)
print("R2:", r2)
````

***Independencia (Prueba de Durbin-Watson)***

Durbin-Watson

Evalúa si los residuos del modelo están autocorrelacionados (es decir, si el error en una observación está relacionado con el error en la siguiente).

Puedes asumir que el supuesto de independencia de los errores se cumple.

***Criterios generales de VIF:***

VIF ≤ 1: No hay multicolinealidad. La variable no está correlacionada con otras.
1 < VIF < 5: Hay una correlación moderada, pero no es preocupante.

VIF ≥ 5: Indica que hay multicolinealidad significativa con otras variables y debe ser revisado.

VIF ≥ 10: Hay una multicolinealidad severa. Es recomendable eliminar o modificar la variable.

***Prueba de Durbin-Watson*** es una prueba estadística que se utiliza para detectar la autocorrelación de los residuos en un modelo de regresión, es decir, si los residuos del modelo están correlacionados con los residuos anteriores. La autocorrelación de los residuos puede violar uno de los supuestos clave de la regresión lineal, que es la independencia de los residuos.

***Interpretación de la estadística de Durbin-Watson:***

La estadística de Durbin-Watson se encuentra en un rango de valores entre 0 y 4:

Durbin-Watson ≈ 2: Esto indica que no hay autocorrelación en los residuos. Los residuos son independientes.

Durbin-Watson < 2: Indica autocorrelación positiva. Esto significa que los residuos tienden a ser similares entre sí, lo que puede implicar que el modelo no ha capturado adecuadamente alguna estructura en los datos.

Durbin-Watson > 2: Indica autocorrelación negativa. Los residuos muestran una tendencia opuesta entre sí.

Durbin-Watson: 0.9999999999999855

Este valor está cerca de 0, lo que indica una autocorrelación positiva fuerte en los residuos.

***Posibles implicaciones de un valor bajo (cerca de 0):***

Autocorrelación positiva: Los residuos están fuertemente correlacionados, lo que puede significar que el modelo no ha capturado toda la estructura en los datos. Este comportamiento puede indicar que falta una variable importante o que el modelo no es lo suficientemente adecuado para los datos.

Problema de independencia: La independencia de los residuos es uno de los supuestos fundamentales de la regresión lineal. Homocedasticidad: Si el gráfico de residuos muestra un patrón en lugar de dispersarse aleatoriamente alrededor de 0, es posible que se necesiten transformaciones en los datos.

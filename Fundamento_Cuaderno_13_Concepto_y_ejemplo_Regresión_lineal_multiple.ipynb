## <font color="rede">13.1 Definiciones

驴Qu茅 es la Regresi贸n Lineal m煤ltiple?
La Regresi贸n Lineal m煤ltiple es un modelo que busca encontrar una relaci贸n lineal entre variables independientes (X) y la variable dependiente (Y).
Vamos a modificar las variables para reprensentar la Regresi贸n Lineal Simple y adoptarla al modelo general m煤ltiple, la siguiente escuaci贸n es igual que la anterior solo que usamos $ \beta$

#### F贸rmula matem谩tica:
$Y = \beta_0 + \beta_1 X + \epsilon$

### Descripci贸n de los t茅rminos:
* **Y**: Variable dependiente.
* **X**: Variable independiente.
* **$beta_0$**: Intersecci贸n (valor de \( Y \) cuando \( X = 0 \)).
* **$beta_1$**: Pendiente (grado de cambio en \( Y \) por unidad de cambio en \( X \)).
* **$epsilon$**: T茅rmino de error.

### <font color="red">13.1.1 Ecuaci贸n con la Regresi贸n Lineal M煤ltiple
Es una extensi贸n de la regresi贸n lineal simple, donde la variable dependiente est谩 relacionada con dos o m谩s variables independientes $(X_1, X_2, ..., X_p)$

F贸rmula matem谩tica:

  $Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \beta_p X_p + \epsilon$

Donde:  
* $Y$: Variable dependiente.  
* $X_1, X_2, \dots, X_p$: Variables independientes.  
* $\beta_0, \beta_1, \dots, \beta_p$: Coeficientes del modelo.

## <font color="rede">13.2 Bondad de Ajuste del Modelo
Error Est谩ndar Residual (RSE)
El Error Est谩ndar Residual mide la desviaci贸n promedio entre los valores observados y los valores predichos por el modelo.

$RSE = \sqrt{\frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{n - p - 1}}$

Donde:  
* $y_i$: Valor observado.  
* $\hat{y}_i$: Valor predicho.  
* $n$: N煤mero de observaciones.  
* $p$: N煤mero de predictores.  


### <font color="red">13.2.1 Coeficiente de Determinaci贸n ($R^2$)
El $R^2$ indica qu茅 proporci贸n de la variabilidad de la variable dependiente es explicada por el modelo.  

$$R^2 = 1 - \frac{\text{Suma de Cuadrados Residuales (RSS)}}{\text{Suma Total de Cuadrados (TSS)}}$$  

Tambi茅n puede calcularse como:  

$$R^2 = \frac{\sum (\hat{y}_i - \bar{y})^2}{\sum (y_i - \bar{y})^2}$$  

* Valor de $R^2$ cercano a 1 indica un buen ajuste.  

#### <font color="red">Significancia de los Predictores  

Se eval煤a la significancia de los predictores usando el valor $p$ de las pruebas de hip贸tesis. Si $p < \alpha$ (nivel de significancia, com煤nmente 0.05), el predictor es significativo.

## <font color="rede">13.4 Validaci贸n del Modelo
Para garantizar la validez de una regresi贸n lineal, se deben verificar las siguientes condiciones:

### <font color="red">13.4.1 No Colinealidad o Multicolinealidad  

La multicolinealidad ocurre cuando los predictores en un modelo de regresi贸n est谩n altamente correlacionados entre s铆, lo que dificulta la estimaci贸n precisa de los coeficientes. En presencia de multicolinealidad, los resultados del modelo pueden ser inestables, y las interpretaciones de los coeficientes pueden volverse poco confiables. Es importante evaluar y detectar la multicolinealidad en modelos de regresi贸n.  

Para evaluar la multicolinealidad, se utilizan dos m茅todos principales:

***ndice de Tolerancia:*** Este 铆ndice mide cu谩nto de la varianza de un predictor puede explicarse por los dem谩s predictores. Se calcula como:

$\text{ndice de Tolerancia} = 1 - R_j^2$

donde $(R_j^2)$ es el coeficiente de determinaci贸n de la regresi贸n del predictor $j$ sobre los otros predictores. Un valor de tolerancia cercano a 0 indica alta multicolinealidad.

**Factor de Inflaci贸n de la Varianza (VIF, por sus siglas en ingl茅s)**: El VIF mide cu谩nta varianza en el coeficiente estimado de un predictor se "infla" debido a la correlaci贸n con otros predictores. Se calcula como:
   
$VIF = \frac{1}{1 - R_j^2}$

donde $(R_j^2)$ es el coeficiente de determinaci贸n de la regresi贸n del predictor $(j)$ sobre los dem谩s predictores. Un VIF alto (t铆picamente mayor a 10) indica una alta multicolinealidad.

En general:
- Si la tolerancia es peque帽a o el VIF es grande, significa que hay multicolinealidad.
- Se recomienda tratar de eliminar o combinar variables altamente correlacionadas para reducir la multicolinealidad y mejorar la estabilidad del modelo.

### <font color="red">13.4.2 Relaci贸n Lineal
La relaci贸n entre los predictores y la variable dependiente debe ser lineal. Puede verificarse mediante:
* Diagramas de dispersi贸n
* Transformaciones logar铆tmicas si es necesario.

### <font color="red">13.4.3 Normalidad de los Residuos
Los residuos deben seguir una distribuci贸n normal. Puede verificarse con:
* Prueba de Shapiro-Wilk
* Gr谩ficos Q-Q

### <font color="red">13.4.4 Homocedasticidad
La **homocedasticidad** se refiere a que la varianza de los residuos debe ser constante a lo largo de los valores ajustados (predicciones). Si la varianza de los residuos cambia con los valores ajustados, se dice que hay **heterocedasticidad**. Para verificar la homocedasticidad, se puede graficar los **residuos versus los valores ajustados**. En una distribuci贸n homog茅nea, los residuos deben estar dispersos aleatoriamente alrededor de cero, sin formar patrones sistem谩ticos. En los ejercicios pr谩cticos vamos a ver la gr谩fica qqplot para verificar la varianza de los residuos.

### <font color="red">13.4.5 No Autocorrelaci贸n
La **autocorrelaci贸n** ocurre cuando los residuos no son independientes entre s铆, es decir, hay correlaci贸n temporal o secuencial entre ellos. Esto es un problema porque violar铆a uno de los supuestos fundamentales de la regresi贸n.

La **prueba de Durbin-Watson** se utiliza para evaluar la autocorrelaci贸n de los residuos. El estad铆stico de Durbin-Watson (\(d\)) se calcula con la siguiente f贸rmula:

$d = \frac{\sum_{t=2}^n (e_t - e_{t-1})^2}{\sum_{t=1}^n e_t^2}$

- $( e_t )$ es el residuo en el tiempo \(t\).
- $( d )$ var铆a entre 0 y 4:
  - $( d \approx 2 )$: No hay autocorrelaci贸n.
  - $( d < 2 )$: Autocorrelaci贸n positiva.
  - $( d > 2 )$: Autocorrelaci贸n negativa.

### <font color="red"><font color="rede">13.4.6 Ausencia de Valores At铆picos o Influyentes
Los valores at铆picos pueden identificarse usando:
* Distancia de Cook
* Leverage

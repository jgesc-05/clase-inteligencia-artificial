```python
# prompt: Crea un arreglo con la variable datos, con datos sintétios,  usnado numpy donde la columna 0 sea edad, columna 1 notas que van de 0 a 5

import numpy as np

# Crear un arreglo con datos sintéticos usando NumPy
datos = np.random.randint(low=0, high=100, size=(100, 2)) # 100 filas, 2 columnas

# Asignar edades a la primera columna (edad entre 18 y 65)
datos[:, 0] = np.random.randint(18, 66, size=100)

# Asignar notas a la segunda columna (notas entre 0 y 5 con un decimal)
datos[:, 1] = np.round(np.random.uniform(0, 5, size=100), 1)

datos

#Cálculo del promedio de la columna 1 del arreglo datos

data_mean = np.mean(datos[:, 1])

data_mean


#la notación de datos[:,:] muestra todo (un split del array en filas y columnas, en realidad una matriz)
#
```

```python
import pandas as pd
import numpy as np
```

```python
# prompt: Crear un df cargado en https://raw.githubusercontent.com/adiacla/bigdata/refs/heads/master/vivienda_clasificacion.csv

url = 'https://raw.githubusercontent.com/adiacla/bigdata/refs/heads/master/vivienda_clasificacion.csv'
df = pd.read_csv(url)

df
```

```python
# prompt: Crea una gráfica de barras para ver el número de habitaciones

import matplotlib.pyplot as plt


# Crear la gráfica de barras
plt.figure(figsize=(10, 6))  # Ajustar el tamaño de la figura
df['habitaciones'].value_counts().plot(kind='bar')
plt.title('Número de habitaciones')
plt.xlabel('Número de habitaciones')
plt.ylabel('Cantidad de viviendas')
plt.xticks(rotation=0)  # Evitar rotación de etiquetas del eje x
plt.show()
```

#Outlier
Un valor atípico (ej: en un grupo de estudiantes, promedio de edad 19, hay un estudiante de 60 años, que es el outlier)

Todo lo que se encuentre arriba del bigote superior es un outlier, que se calcula con la fórmula de Cuartil3*Ir1.5


```python
# prompt: Borrar registros superiores a 32 y dejarlo en el mismo df

# Eliminar registros donde 'habitaciones' es mayor que 32
df = df[df['habitaciones'] <= 32]
```

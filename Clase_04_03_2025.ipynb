```python
#Juego dataset
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from joblib import dump
from sklearn.ensemble import RandomForestClassifier

data = {
    "Género" : ["M", "F", "F", "F", "F", "M", "M", "M", "M", "M", "M"],
    "Edad" : [19,19,18,20,22,21,19,19,23,18,25],
    "Peso" : [80, 52, 55, 60, 62, 68, 72, 60, 68, 57, 80],
    "Altura" : [194,160,155,160,176,185,178,170,175,175,175],
    "Talla" : [45,38,35,37,36,41,41,40,40,40,42]
}

df = pd.DataFrame(data)
df
```

```python
encoded_gender = LabelEncoder()
df["Género"] = encoded_gender.fit_transform(df["Género"])
df
```

```python
dump(encoded_gender, "gender_encoder.joblib")
```

```python
encoded_gender.classes_
```

```python
X = df[["Edad", "Peso", "Altura", "Talla"]]
y = df["Género"]
```

```python
#Entranando el modelo

model = RandomForestClassifier()
model.fit(X,y)
```

```python
dump(model, "random_forest_model.joblib")
print("Modelo guardado correctamente")

#Se hizo un ejercicio llamado app2 en vscode
```
